###  1. Get the list of topic from the topics page

```python
import requests as rq
from bs4 import BeautifulSoup as bs
import pandas as pd
import os

def scrape_topics():
    
    ##################-Topic's Extraction-#############################################################################
    topic_url="https://github.com/topics"
    response=rq.get(topic_url) 
    if response.status_code!=200:
        raise Exception("Failed to load page {}".format(topic_url))
        
    ### parsing HTML
    page_content=response.text
    topic_soup=bs(page_content, "html.parser")
    global doc
    doc= topic_soup
        
    # Creating a dictionary of created data.
    Topic_dict={"Topics": get_topics_titles(doc), "Description":get_description(doc), 
                "Topics_link": get_topics_url(doc)}
    
#   Creating a data frame with Topic_dict
    Topic_df=pd.DataFrame(Topic_dict)
    new_index=list(range(1,31))
    Topic_df.index=new_index
    Topic_df.index.name="Sn.No:"
    return Topic_df                   
    
def get_topics_titles(doc):
    ## Extracting of p tag cointaining specific class attribute topic
        
    topic_title_tags=doc.find_all("p",{"class":"f3 lh-condensed mb-0 mt-1 Link--primary"})
    topic_title_tags_list=[i.text for i in topic_title_tags]
    Topics=topic_title_tags_list
    return Topics
        
def get_description(doc):
     ## Extracting of topic description tags
        
    topic_descr_tags=doc.find_all("p",{"class":"f5 color-fg-muted mb-0 mt-1"})
    topic_descr_tags_list=[(i.text).strip("\n " ) for i in topic_descr_tags]
    Description=topic_descr_tags_list
    return Description
    
def get_topics_url(doc):
       
    ### Extracting Topics Url
    topic_link_tags=doc.find_all("a",{"class":"no-underline flex-1 d-flex flex-column"})
    topic_link_tags_list=["https://github.com"+i["href"] for i in topic_link_tags]
    Topics_link=topic_link_tags_list
    return Topics_link
        

        
        
topics_df= scrape_topics()       
topics_df.head(1)
```
